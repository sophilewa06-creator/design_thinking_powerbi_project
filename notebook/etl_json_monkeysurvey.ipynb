```python
#1 Carga de datos (se utiliza utf-8 por ser respuestas en español)
import json
import pandas as pd

with open("survey_response_mock.json","r", encoding="utf-8") as file:
  data=json.load(file)

responses=data["responses"]

#2 Transformación del JSON
df=pd.json_normalize(responses)

df=df.rename(columns={"demographics.age": "edad",
    "demographics.gender": "genero",
    "demographics.appointment_type": "tipo_cita",
    "demographics.booking_channel": "Canal_Agendamiento",
    "experience.pre_appointment_satisfaction": "satisfaccion_precita(1-5)",
    "experience.post_appointment_satisfaction": "satisfaccion_postcita(1-5)",
    "experience.nps_score": "nps",
    "experience.waiting_time_minutes": "tiempo_espera"})

#3 Exportación para Power BI
df.to_csv("survey_cleaned_for_powerbi.csv", index=False)


```
